```python
import json
import re
import pickle
```

```python
documents=[]
with open('resume_dataset.pickle','rb') as resume:
    df=pickle.load(resume)
    data=df.loc[df['class'] == 'Skills']
    for line in data['text']:
         documents.extend(re.sub(r'\((.)*([year]|[years])(.)*\)','',line.lower()).split(','))
print(len(documents))
with open('data/resume/dataset_Skills.json','r') as skills:
    data=json.load(skills)
    for i in range(0,1450):
        documents.extend(re.sub(r'\((.)*([year]|[years])(.)*\)','',data['0'][str(i)].lower()).split(','))
print(len(documents))
dict_skills=[] 
for skill in documents:
    dict_skills.append(skill.replace('\n       \n\n\n\n\n\n',' ').strip())
d=[]
for s in dict_skills:
    if dict_skills.count(s)>10:
        d.append(s)
print(len(set(d)))
print(set(d))
    
    
```

```python
with open('skills.pickle','wb') as f:
    pickle.dump(d,f)
```
